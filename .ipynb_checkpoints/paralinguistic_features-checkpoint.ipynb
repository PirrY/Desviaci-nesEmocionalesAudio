# Extracción de Características Paralingüísticas de Llamadas Telefónicas

Este notebook implementa una prueba de concepto para:
1. **Diarización de hablantes**: separar el audio de una llamada en segmentos por persona
2. **Extracción de características paralingüísticas**: analizar prosodia, energía, ritmo, etc. por hablante

### Características paralingüísticas que extraemos
| Característica | Descripción |
|---|---|
| Pitch (F0) | Frecuencia fundamental — entonación y emoción |
| Energía (RMS) | Volumen / intensidad de voz |
| Tempo / Rate of Speech | Velocidad al hablar |
| Jitter | Variación ciclo a ciclo del pitch (roughness) |
| Shimmer | Variación de amplitud ciclo a ciclo |
| HNR | Harmonic-to-Noise Ratio — calidad de voz |
| MFCCs | Características espectrales del tracto vocal |
| ZCR | Zero Crossing Rate — indica fricativas/silencios |
| Duración de pausas | Silencios entre habla |

## 2. Cargar el archivo de audio

Soportamos los formatos más comunes de llamadas: `.wav`, `.mp3`, `.ogg`, `.m4a`.

pyannote requiere **mono, 16kHz** — convertimos automáticamente si es necesario.

```python
# =============================================
# CONFIGURACIÓN: cambia esta ruta a tu archivo
# =============================================
AUDIO_PATH = "llamada_ejemplo.wav"  # <-- cambia aquí
HF_TOKEN = "hf_xxxxxxxxxxxxxxxxxxxx"  # <-- tu token de HuggingFace
OUTPUT_DIR = Path("output_speakers")
OUTPUT_DIR.mkdir(exist_ok=True)

# Tasa de muestreo estándar para análisis
TARGET_SR = 16000


def load_audio(path: str, target_sr: int = 16000):
    """
    Carga un archivo de audio y lo convierte a mono/16kHz.
    Devuelve (waveform_numpy, sample_rate).
    """
    audio = AudioSegment.from_file(path)
    # Convertir a mono
    audio = audio.set_channels(1)
    # Convertir sample rate
    audio = audio.set_frame_rate(target_sr)

    # Guardar versión procesada para pyannote
    processed_path = str(OUTPUT_DIR / "audio_processed.wav")
    audio.export(processed_path, format="wav")

    # Cargar con librosa para análisis
    y, sr = librosa.load(processed_path, sr=target_sr, mono=True)
    duration = librosa.get_duration(y=y, sr=sr)

    print(f"Audio cargado:")
    print(f"  Duración  : {duration:.2f} s ({duration/60:.1f} min)")
    print(f"  Sample rate: {sr} Hz")
    print(f"  Samples   : {len(y):,}")
    print(f"  Guardado en: {processed_path}")

    return y, sr, processed_path


y_full, sr, processed_audio_path = load_audio(AUDIO_PATH, TARGET_SR)
```

## 3. Diarización de hablantes

Usamos **pyannote/speaker-diarization-3.1**, que internamente combina:
- Detección de actividad de voz (VAD) para encontrar segmentos con habla
- Segmentación en bloques homogéneos de hablante
- Clustering para agrupar segmentos del mismo hablante

El resultado es una **anotación de tiempo** con etiquetas como `SPEAKER_00`, `SPEAKER_01`, etc.

```python
def run_diarization(audio_path: str, hf_token: str, num_speakers: int = 2):
    """
    Ejecuta diarización con pyannote.
    num_speakers=2 porque son llamadas entre 2 personas.
    Devuelve la anotación de diarización.
    """
    print("Cargando pipeline de diarización (puede tardar la primera vez)...")
    pipeline = Pipeline.from_pretrained(
        "pyannote/speaker-diarization-3.1",
        use_auth_token=hf_token
    )

    print(f"Diarizando '{audio_path}'...")
    diarization = pipeline(
        audio_path,
        num_speakers=num_speakers  # fijamos 2 porque sabemos que son 2 personas
    )

    # Parsear segmentos a DataFrame
    segments = []
    for turn, _, speaker in diarization.itertracks(yield_label=True):
        segments.append({
            'speaker': speaker,
            'start': turn.start,
            'end': turn.end,
            'duration': turn.end - turn.start
        })

    df = pd.DataFrame(segments)
    print(f"\nDiarización completada: {len(df)} segmentos encontrados")
    print(df.groupby('speaker')['duration'].agg(['count', 'sum', 'mean']).round(2))

    return diarization, df


diarization, df_segments = run_diarization(processed_audio_path, HF_TOKEN, num_speakers=2)
```

### Visualizar la línea de tiempo de diarización

```python
def plot_diarization(df_segments, audio_duration):
    speakers = df_segments['speaker'].unique()
    colors = {spk: c for spk, c in zip(speakers, ['#2196F3', '#FF5722', '#4CAF50', '#9C27B0'])}

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 5), sharex=True,
                                    gridspec_kw={'height_ratios': [3, 1]})

    # Forma de onda con color por hablante
    times = np.linspace(0, len(y_full) / sr, num=len(y_full))
    ax1.plot(times, y_full, color='lightgrey', linewidth=0.3, zorder=1)

    for _, row in df_segments.iterrows():
        mask = (times >= row['start']) & (times <= row['end'])
        ax1.fill_between(times, y_full, where=mask,
                         color=colors[row['speaker']], alpha=0.5, zorder=2)

    ax1.set_ylabel('Amplitud')
    ax1.set_title('Forma de onda coloreada por hablante')
    patches = [mpatches.Patch(color=colors[s], label=s) for s in speakers]
    ax1.legend(handles=patches, loc='upper right')

    # Línea de tiempo tipo gantt
    for i, spk in enumerate(speakers):
        segs = df_segments[df_segments['speaker'] == spk]
        for _, row in segs.iterrows():
            ax2.barh(i, row['duration'], left=row['start'],
                     color=colors[spk], height=0.6, alpha=0.85)

    ax2.set_yticks(range(len(speakers)))
    ax2.set_yticklabels(speakers)
    ax2.set_xlabel('Tiempo (s)')
    ax2.set_title('Línea de tiempo de turnos de habla')

    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / 'diarization_timeline.png', dpi=150, bbox_inches='tight')
    plt.show()


audio_duration = len(y_full) / sr
plot_diarization(df_segments, audio_duration)
```

## 4. Extraer audio por hablante

Concatenamos todos los segmentos de cada hablante para generar un audio limpio por persona.
Esto nos permite analizar cada hablante de forma independiente.

```python
def extract_speaker_audio(y_full: np.ndarray, sr: int, df_segments: pd.DataFrame):
    """
    Para cada hablante, concatena sus segmentos de audio en una sola señal.
    Retorna dict {speaker_id: np.ndarray}
    """
    speaker_audio = {}

    for speaker in df_segments['speaker'].unique():
        segs = df_segments[df_segments['speaker'] == speaker]
        chunks = []

        for _, row in segs.iterrows():
            start_sample = int(row['start'] * sr)
            end_sample = int(row['end'] * sr)
            chunk = y_full[start_sample:end_sample]
            # Filtrar segmentos muy cortos (< 0.1 s) — suelen ser ruido
            if len(chunk) > sr * 0.1:
                chunks.append(chunk)

        if chunks:
            audio_concat = np.concatenate(chunks)
            speaker_audio[speaker] = audio_concat

            # Guardar como archivo WAV
            out_path = OUTPUT_DIR / f"{speaker}.wav"
            sf.write(str(out_path), audio_concat, sr)

            total_sec = len(audio_concat) / sr
            print(f"{speaker}: {len(chunks)} segmentos, {total_sec:.1f} s de audio ({total_sec/60:.2f} min)")
            print(f"  -> Guardado en {out_path}")

    return speaker_audio


speaker_audio = extract_speaker_audio(y_full, sr, df_segments)
speakers = list(speaker_audio.keys())
```

## 5. Extracción de características paralingüísticas

### 5.1 Funciones de extracción

Definimos funciones modulares para cada grupo de características.

```python
# ─────────────────────────────────────────────────────────
# Pitch / F0  (con librosa)
# ─────────────────────────────────────────────────────────

def extract_pitch_features(y: np.ndarray, sr: int) -> dict:
    """
    Extrae estadísticas del pitch (F0) usando pyin (probabilistic YIN).
    Solo analiza frames con voz detectada (voiced_flag=True).
    """
    f0, voiced_flag, voiced_probs = librosa.pyin(
        y,
        fmin=librosa.note_to_hz('C2'),   # ~65 Hz  — mínimo realista de voz humana
        fmax=librosa.note_to_hz('C7'),   # ~2093 Hz
        sr=sr
    )

    voiced_f0 = f0[voiced_flag]  # descartamos frames sin voz

    if len(voiced_f0) == 0:
        return {k: np.nan for k in ['pitch_mean', 'pitch_std', 'pitch_min',
                                     'pitch_max', 'pitch_range', 'pitch_voiced_ratio']}

    return {
        'pitch_mean':         np.mean(voiced_f0),
        'pitch_std':          np.std(voiced_f0),
        'pitch_min':          np.min(voiced_f0),
        'pitch_max':          np.max(voiced_f0),
        'pitch_range':        np.max(voiced_f0) - np.min(voiced_f0),
        'pitch_voiced_ratio': voiced_flag.mean(),  # fracción del tiempo con voz
    }


# ─────────────────────────────────────────────────────────
# Energía / Intensidad
# ─────────────────────────────────────────────────────────

def extract_energy_features(y: np.ndarray, sr: int) -> dict:
    """
    RMS energy y sus estadísticas. Refleja el volumen/intensidad.
    """
    rms = librosa.feature.rms(y=y)[0]
    # Convertir a dB para interpretación más natural
    rms_db = librosa.amplitude_to_db(rms, ref=np.max)

    return {
        'energy_mean_db':   np.mean(rms_db),
        'energy_std_db':    np.std(rms_db),
        'energy_max_db':    np.max(rms_db),
        'energy_dynamic_range': np.max(rms_db) - np.min(rms_db),
    }


# ─────────────────────────────────────────────────────────
# MFCCs  (características espectrales del tracto vocal)
# ─────────────────────────────────────────────────────────

def extract_mfcc_features(y: np.ndarray, sr: int, n_mfcc: int = 13) -> dict:
    """
    13 coeficientes MFCC + delta + delta-delta.
    Cada coeficiente captura una dimensión del espectro vocal.
    """
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    delta = librosa.feature.delta(mfcc)
    delta2 = librosa.feature.delta(mfcc, order=2)

    features = {}
    for i in range(n_mfcc):
        features[f'mfcc{i+1}_mean'] = np.mean(mfcc[i])
        features[f'mfcc{i+1}_std']  = np.std(mfcc[i])

    features['mfcc_delta_mean']  = np.mean(delta)
    features['mfcc_delta2_mean'] = np.mean(delta2)

    return features


# ─────────────────────────────────────────────────────────
# Zero Crossing Rate  (ritmo y sonoridad)
# ─────────────────────────────────────────────────────────

def extract_zcr_features(y: np.ndarray) -> dict:
    """
    ZCR alto → fricativas, consonantes sordas, ruido.
    ZCR bajo → vocales, sonidos sonoros.
    """
    zcr = librosa.feature.zero_crossing_rate(y)[0]
    return {
        'zcr_mean': np.mean(zcr),
        'zcr_std':  np.std(zcr),
    }


# ─────────────────────────────────────────────────────────
# Jitter, Shimmer, HNR  (con Praat via parselmouth)
# ─────────────────────────────────────────────────────────

def extract_praat_features(y: np.ndarray, sr: int,
                            f0_min: float = 75, f0_max: float = 500) -> dict:
    """
    Jitter: variación de periodo a periodo del pitch
      → valores altos indican voz tensa/ronca
    Shimmer: variación de amplitud de ciclo a ciclo
      → valores altos indican breathiness o patología
    HNR: relación armónicos/ruido
      → valores bajos = voz más ruidosa/breathy
    """
    # parselmouth requiere float64
    y64 = y.astype(np.float64)
    snd = parselmouth.Sound(y64, sampling_frequency=sr)

    # Point process para jitter/shimmer
    point_process = call(snd, "To PointProcess (periodic, cc)", f0_min, f0_max)

    jitter_local = call(point_process, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
    jitter_rap   = call(point_process, "Get jitter (rap)",   0, 0, 0.0001, 0.02, 1.3)
    shimmer_local = call([snd, point_process], "Get shimmer (local)",   0, 0, 0.0001, 0.02, 1.3, 1.6)
    shimmer_apq3  = call([snd, point_process], "Get shimmer (apq3)",    0, 0, 0.0001, 0.02, 1.3, 1.6)

    # HNR
    harmonicity = call(snd, "To Harmonicity (cc)", 0.01, f0_min, 0.1, 1.0)
    hnr = call(harmonicity, "Get mean", 0, 0)

    return {
        'jitter_local':   jitter_local,
        'jitter_rap':     jitter_rap,
        'shimmer_local':  shimmer_local,
        'shimmer_apq3':   shimmer_apq3,
        'hnr_mean_db':    hnr,
    }


# ─────────────────────────────────────────────────────────
# Pausas y ritmo de habla
# ─────────────────────────────────────────────────────────

def extract_speech_rate_features(y: np.ndarray, sr: int,
                                  df_speaker_segs: pd.DataFrame) -> dict:
    """
    Analiza el ritmo de habla a partir de los segmentos del hablante:
    - Duración promedio de turno
    - Número de turnos
    - Densidad de pausas (tiempo en silencio vs. tiempo total asignado)
    """
    durations = df_speaker_segs['duration'].values
    total_speech = durations.sum()

    # Estimar sílabas por segundo usando onset detection como proxy
    # (onset = inicio de sílaba de forma aproximada)
    onsets = librosa.onset.onset_detect(y=y, sr=sr, units='time')
    syllable_rate = len(onsets) / total_speech if total_speech > 0 else 0

    return {
        'n_turns':           len(durations),
        'total_speech_s':    total_speech,
        'mean_turn_s':       np.mean(durations),
        'std_turn_s':        np.std(durations),
        'max_turn_s':        np.max(durations),
        'syllable_rate_hz':  syllable_rate,   # sílabas/seg (aprox)
    }


print("Funciones de extracción definidas.")
```

### 5.2 Extraer todas las características para cada hablante

```python
def extract_all_features(speaker_audio: dict, sr: int, df_segments: pd.DataFrame) -> pd.DataFrame:
    """
    Recorre todos los hablantes y combina todas las características en un DataFrame.
    """
    all_features = []

    for speaker, y in speaker_audio.items():
        print(f"\nExtrayendo features para {speaker}...")
        feats = {'speaker': speaker}

        print("  Pitch...", end=' ')
        feats.update(extract_pitch_features(y, sr))
        print("OK")

        print("  Energía...", end=' ')
        feats.update(extract_energy_features(y, sr))
        print("OK")

        print("  MFCCs...", end=' ')
        feats.update(extract_mfcc_features(y, sr))
        print("OK")

        print("  ZCR...", end=' ')
        feats.update(extract_zcr_features(y))
        print("OK")

        print("  Jitter/Shimmer/HNR (Praat)...", end=' ')
        try:
            feats.update(extract_praat_features(y, sr))
            print("OK")
        except Exception as e:
            print(f"ADVERTENCIA: {e}")
            for k in ['jitter_local', 'jitter_rap', 'shimmer_local', 'shimmer_apq3', 'hnr_mean_db']:
                feats[k] = np.nan

        print("  Ritmo de habla...", end=' ')
        df_spk = df_segments[df_segments['speaker'] == speaker]
        feats.update(extract_speech_rate_features(y, sr, df_spk))
        print("OK")

        all_features.append(feats)

    return pd.DataFrame(all_features).set_index('speaker')


df_features = extract_all_features(speaker_audio, sr, df_segments)
print("\n=== Extracción completada ===")
df_features
```

### 6.3 Spectrograma y energía por hablante

```python
for speaker, y in speaker_audio.items():
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 7), sharex=True,
                                    gridspec_kw={'height_ratios': [3, 1]})

    # Spectrograma mel
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
    S_db = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_db, x_axis='time', y_axis='mel',
                                    sr=sr, fmax=8000, ax=ax1, cmap='magma')
    fig.colorbar(img, ax=ax1, format='%+2.0f dB')
    ax1.set_title(f'Espectrograma Mel — {speaker}')

    # Energía RMS
    rms = librosa.feature.rms(y=y)[0]
    times_rms = librosa.times_like(rms, sr=sr)
    ax2.fill_between(times_rms, librosa.amplitude_to_db(rms, ref=np.max),
                     alpha=0.8, color='#FF5722')
    ax2.set_ylabel('dB')
    ax2.set_xlabel('Tiempo (s)')
    ax2.set_title('Energía RMS')

    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / f'spectrogram_{speaker}.png', dpi=150, bbox_inches='tight')
    plt.show()
```

## 7. Análisis por ventanas de tiempo (evolución dinámica)

En lugar de un único vector de features por hablante, podemos calcular
las características en ventanas deslizantes para ver **cómo evolucionan a lo largo de la llamada**.

```python
def windowed_features(y_full: np.ndarray, sr: int, df_segments: pd.DataFrame,
                       window_s: float = 5.0, hop_s: float = 2.5) -> pd.DataFrame:
    """
    Calcula pitch y energía en ventanas de `window_s` segundos
    avanzando `hop_s` segundos, respetando qué hablante está activo.
    """
    rows = []
    total_duration = len(y_full) / sr
    t = 0.0

    while t + window_s <= total_duration:
        t_end = t + window_s

        # Determinar hablante dominante en esta ventana
        overlap = df_segments.copy()
        overlap['overlap'] = overlap.apply(
            lambda r: max(0, min(r['end'], t_end) - max(r['start'], t)), axis=1
        )
        dom = overlap.loc[overlap['overlap'].idxmax()]

        # Extraer señal de la ventana
        s_start = int(t * sr)
        s_end   = int(t_end * sr)
        y_win   = y_full[s_start:s_end]

        # Pitch
        f0, voiced, _ = librosa.pyin(y_win,
                                      fmin=librosa.note_to_hz('C2'),
                                      fmax=librosa.note_to_hz('C7'), sr=sr)
        voiced_f0 = f0[voiced]
        pitch_mean = np.nanmean(voiced_f0) if len(voiced_f0) > 0 else np.nan

        # Energía
        rms = librosa.feature.rms(y=y_win)[0]
        energy_mean = librosa.amplitude_to_db(rms, ref=np.max).mean()

        rows.append({
            'time_s':      t,
            'speaker':     dom['speaker'],
            'pitch_mean':  pitch_mean,
            'energy_mean_db': energy_mean,
        })
        t += hop_s

    return pd.DataFrame(rows)


df_windowed = windowed_features(y_full, sr, df_segments, window_s=5.0, hop_s=2.5)
df_windowed.head()
```

```python
# Visualizar evolución del pitch y energía a lo largo de la llamada
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 7), sharex=True)
colors_map = dict(zip(speakers, ['#2196F3', '#FF5722']))

for spk in speakers:
    d = df_windowed[df_windowed['speaker'] == spk]
    ax1.plot(d['time_s'], d['pitch_mean'], 'o-', color=colors_map[spk],
             label=spk, markersize=4, linewidth=1.5, alpha=0.85)
    ax2.plot(d['time_s'], d['energy_mean_db'], 's-', color=colors_map[spk],
             label=spk, markersize=4, linewidth=1.5, alpha=0.85)

ax1.set_ylabel('Pitch medio (Hz)')
ax1.set_title('Evolución del Pitch a lo largo de la llamada')
ax1.legend()

ax2.set_ylabel('Energía media (dB)')
ax2.set_xlabel('Tiempo de llamada (s)')
ax2.set_title('Evolución de la Energía a lo largo de la llamada')
ax2.legend()

plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'temporal_evolution.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 9. Guía de interpretación

### ¿Qué nos dice cada característica?

| Característica | Valor bajo | Valor alto | Aplicación |
|---|---|---|---|
| **pitch_mean** | Voz grave (hombre típico: ~120 Hz) | Voz aguda (mujer típica: ~220 Hz) | Identificación de género, detección de estrés |
| **pitch_std** | Voz monótona, plana | Voz muy expresiva/melodiosa | Estado emocional, engagement |
| **energy_mean_db** | Voz suave/baja | Voz fuerte/alta | Confianza, asertividad |
| **jitter_local** | Voz regular y clara (< 1%) | Voz tensa, ronca o patológica (> 1%) | Calidad vocal, estrés |
| **shimmer_local** | Voz estable (< 3%) | Voz breathy o inestable (> 3%) | Calidad vocal, emoción |
| **hnr_mean_db** | Voz ruidosa/breathiness (< 10 dB) | Voz limpia y tonal (> 20 dB) | Calidad de voz, salud vocal |
| **syllable_rate_hz** | Habla lenta (< 3 síl/s) | Habla rápida (> 6 síl/s) | Nerviosismo, fluidez |
| **mean_turn_s** | Turnos cortos (reactivo) | Turnos largos (dominante) | Dinámica conversacional |
