# Detección de Desviaciones Emocionales en Llamadas

## Objetivo
Dado el audio de una llamada entre dos personas:
1. Separar los hablantes
2. Calcular una **línea base paralingüística** de cada persona (cómo habla cuando está neutral)
3. Medir **desviaciones** de esa base a lo largo del tiempo
4. Identificar **momentos de activación emocional** (enojo, excitación, estrés)

## Fundamento teórico

Las emociones de alta activación (enojo, excitación, miedo) producen cambios medibles y consistentes:

| Característica | Neutral → Enojo/Excitación | Neutral → Tristeza/Calma |
|---|---|---|
| Pitch medio (F0) | ↑ sube | ↓ baja |
| Variación de pitch | ↑ aumenta | ↓ disminuye |
| Energía (dB) | ↑ sube | ↓ baja |
| Tasa de habla | ↑ acelera | ↓ desacelera |
| Jitter | ↑ aumenta (voz tensa) | variable |
| HNR | ↓ baja (más ruido) | — |

La estrategia: calcular un **z-score compuesto** = cuántas desviaciones estándar se aleja cada ventana de la base de esa persona.

---

## Pipeline
```
Audio de llamada
      ↓
 Diarización → Segmentos por hablante
      ↓
 Features en ventanas deslizantes (5s hop 1s)
      ↓
 Línea base: primeros N segundos (o percentil 25-75)
      ↓
 Z-score por feature → Score compuesto de desviación
      ↓
 Umbral adaptativo → Detección de momentos clave
      ↓
 Visualización del arco emocional
```

````python
import numpy as np
import pandas as pd
import librosa
import librosa.display
import soundfile as sf
import parselmouth
from parselmouth.praat import call
from pydub import AudioSegment
from pyannote.audio import Pipeline
from scipy.ndimage import uniform_filter1d
from scipy.stats import zscore
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 100
sns.set_style('whitegrid')

print('OK')
````

## 2. Parámetros globales

Todo lo configurable está aquí para fácil ajuste.

````python
# ─── ENTRADA ────────────────────────────────────────────────
AUDIO_PATH  = "llamada_ejemplo.wav"   # archivo de la llamada
HF_TOKEN    = "hf_xxxxxxxxxxxxxxxxxxxx"  # token HuggingFace
OUTPUT_DIR  = Path("output_emotional")
OUTPUT_DIR.mkdir(exist_ok=True)

# ─── AUDIO ──────────────────────────────────────────────────
TARGET_SR   = 16_000   # Hz — requerido por pyannote

# ─── VENTANA DESLIZANTE ─────────────────────────────────────
WINDOW_S    = 5.0      # duración de cada ventana de análisis (segundos)
HOP_S       = 1.0      # paso entre ventanas (solapamiento implícito de 4 s)

# ─── LÍNEA BASE ─────────────────────────────────────────────
# Estrategia 'first_n': usa los primeros N segundos de habla del hablante
# Estrategia 'iqr':     usa el rango intercuartil de toda la llamada (más robusto)
BASELINE_STRATEGY = 'first_n'   # 'first_n' | 'iqr'
BASELINE_FIRST_N_S = 60         # segundos de habla usados como base

# ─── DETECCIÓN ──────────────────────────────────────────────
# Features y sus pesos para el score compuesto
# Mayor peso → más influencia en el score final
FEATURE_WEIGHTS = {
    'pitch_mean':     1.5,   # pitch sube mucho en emociones de alta activación
    'pitch_std':      1.2,   # variación de pitch = expresividad emocional
    'energy_db':      1.3,   # energía/volumen
    'speech_rate':    1.0,   # tasa de sílabas
    'jitter':         0.8,   # calidad de voz (más ruidosa bajo estrés)
    'hnr':            0.7,   # harmonic-to-noise ratio
}

# Umbral para declarar un momento como "emocionalmente activo"
# (en z-scores del score compuesto)
ALERT_THRESHOLD = 2.0    # desviaciones estándar sobre la base

SPEAKER_COLORS = ['#2196F3', '#FF5722']  # azul, naranja
````

## 3. Carga de audio y diarización

````python
def load_and_prepare(audio_path: str, target_sr: int = 16_000):
    """Carga, convierte a mono/16kHz y devuelve (y, sr, processed_path)."""
    audio = AudioSegment.from_file(audio_path)
    audio = audio.set_channels(1).set_frame_rate(target_sr)
    processed = str(OUTPUT_DIR / "audio_16k_mono.wav")
    audio.export(processed, format="wav")
    y, sr = librosa.load(processed, sr=target_sr, mono=True)
    duration = len(y) / sr
    print(f"Audio: {duration:.1f} s ({duration/60:.2f} min) @ {sr} Hz")
    return y, sr, processed


def diarize(audio_path: str, hf_token: str, num_speakers: int = 2) -> pd.DataFrame:
    """Corre pyannote y devuelve DataFrame con columnas [speaker, start, end, duration]."""
    print("Cargando pipeline de diarización...")
    pipeline = Pipeline.from_pretrained(
        "pyannote/speaker-diarization-3.1",
        use_auth_token=hf_token
    )
    print("Diarizando...")
    diarization = pipeline(audio_path, num_speakers=num_speakers)

    rows = []
    for turn, _, speaker in diarization.itertracks(yield_label=True):
        rows.append(dict(speaker=speaker, start=turn.start,
                         end=turn.end, duration=turn.end - turn.start))
    df = pd.DataFrame(rows)
    print(f"\n{len(df)} segmentos detectados")
    print(df.groupby('speaker')['duration'].agg(['count','sum']).rename(
        columns={'count': 'turnos', 'sum': 'seg_totales'}).round(1))
    return df


y_full, sr, processed_path = load_and_prepare(AUDIO_PATH, TARGET_SR)
df_segments = diarize(processed_path, HF_TOKEN, num_speakers=2)
speakers = df_segments['speaker'].unique().tolist()
color_map = {spk: c for spk, c in zip(speakers, SPEAKER_COLORS)}
````

## 4. Extracción de features en ventanas deslizantes

Para cada ventana de `WINDOW_S` segundos calculamos un vector de características.
La ventana se etiqueta con el hablante que más tiempo ocupa en ese intervalo.

**¿Por qué ventanas deslizantes y no segmentos?**  
Los segmentos de diarización son muy variables en duración (0.2 s a 30 s).
Las ventanas fijas garantizan que todas las estimaciones tienen el mismo contexto acústico,
lo que hace comparables los z-scores.

````python
def features_for_window(y_win: np.ndarray, sr: int) -> dict:
    """
    Extrae el vector de features para una ventana de audio.
    Devuelve NaN en features que no se pueden calcular (ventana sin voz, etc.)
    """
    feats = {}

    # ── Pitch ────────────────────────────────────────────────
    f0, voiced, _ = librosa.pyin(
        y_win,
        fmin=librosa.note_to_hz('C2'),
        fmax=librosa.note_to_hz('C7'),
        sr=sr
    )
    voiced_f0 = f0[voiced]
    if len(voiced_f0) >= 5:   # mínimo de frames para estimar
        feats['pitch_mean'] = np.mean(voiced_f0)
        feats['pitch_std']  = np.std(voiced_f0)
        feats['voiced_ratio'] = voiced.mean()
    else:
        feats['pitch_mean'] = np.nan
        feats['pitch_std']  = np.nan
        feats['voiced_ratio'] = 0.0

    # ── Energía ──────────────────────────────────────────────
    rms = librosa.feature.rms(y=y_win)[0]
    feats['energy_db'] = librosa.amplitude_to_db(rms, ref=1.0).mean()

    # ── Tasa de habla (onsets como proxy de sílabas) ──────────
    onsets = librosa.onset.onset_detect(y=y_win, sr=sr, units='time')
    duration = len(y_win) / sr
    feats['speech_rate'] = len(onsets) / duration if duration > 0 else 0.0

    # ── Jitter y HNR (Praat) ─────────────────────────────────
    try:
        y64 = y_win.astype(np.float64)
        snd = parselmouth.Sound(y64, sampling_frequency=sr)
        pp  = call(snd, "To PointProcess (periodic, cc)", 75, 500)

        jitter = call(pp, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
        feats['jitter'] = jitter if np.isfinite(jitter) else np.nan

        harm = call(snd, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
        hnr  = call(harm, "Get mean", 0, 0)
        feats['hnr'] = hnr if np.isfinite(hnr) else np.nan
    except Exception:
        feats['jitter'] = np.nan
        feats['hnr']    = np.nan

    return feats


def dominant_speaker(t_start: float, t_end: float, df_seg: pd.DataFrame) -> str:
    """Retorna el hablante con más tiempo dentro del intervalo [t_start, t_end]."""
    overlap = df_seg.copy()
    overlap['ov'] = overlap.apply(
        lambda r: max(0, min(r['end'], t_end) - max(r['start'], t_start)), axis=1
    )
    if overlap['ov'].max() == 0:
        return 'SILENCE'
    return overlap.loc[overlap['ov'].idxmax(), 'speaker']


def extract_windowed_features(y: np.ndarray, sr: int, df_seg: pd.DataFrame,
                               window_s: float, hop_s: float) -> pd.DataFrame:
    """
    Recorre el audio con ventanas deslizantes y extrae features.
    Descarta ventanas de silencio.
    """
    total_s  = len(y) / sr
    rows     = []
    t        = 0.0
    n_wins   = int((total_s - window_s) / hop_s) + 1
    print(f"Procesando {n_wins} ventanas de {window_s}s (hop {hop_s}s)...")

    for i in range(n_wins):
        t_end = t + window_s
        spk   = dominant_speaker(t, t_end, df_seg)

        if spk != 'SILENCE':
            s_start = int(t * sr)
            s_end   = int(t_end * sr)
            y_win   = y[s_start:s_end]
            feats   = features_for_window(y_win, sr)
            feats['time_s']  = t
            feats['speaker'] = spk
            rows.append(feats)

        t += hop_s
        if (i + 1) % 50 == 0:
            print(f"  {i+1}/{n_wins} ventanas procesadas...")

    df = pd.DataFrame(rows)
    print(f"Total: {len(df)} ventanas con habla detectada")
    return df


df_wins = extract_windowed_features(y_full, sr, df_segments, WINDOW_S, HOP_S)
df_wins.head()
````

## 5. Cálculo de la línea base

### Dos estrategias disponibles

**`first_n`** — Usa los primeros N segundos de habla de cada persona.
- Ventaja: captura el estado inicial (típicamente más neutro al inicio de una llamada).
- Riesgo: si la llamada empieza ya tensa, la base queda mal calibrada.

**`iqr`** — Usa la mediana y el IQR de toda la llamada.
- Ventaja: robusto a outliers y picos emocionales cortos.
- Riesgo: si la emoción dura más de la mitad de la llamada, la base se contamina.

> En producción, la mejor base sería una llamada de referencia de esa misma persona en un contexto conocidamente neutro.

````python
FEATURE_COLS = list(FEATURE_WEIGHTS.keys())


def compute_baseline(df_wins: pd.DataFrame, speaker: str,
                     strategy: str, first_n_s: float) -> dict:
    """
    Calcula media y std de referencia para cada feature del hablante.
    Devuelve {'feature': {'mean': ..., 'std': ...}}
    """
    df_spk = df_wins[df_wins['speaker'] == speaker].copy()

    if strategy == 'first_n':
        # Ventanas donde la posición temporal acumulada de habla < first_n_s
        # Ordenamos por tiempo y tomamos hasta alcanzar N segundos de habla
        df_spk = df_spk.sort_values('time_s')
        cumulative_speech = (df_spk.index.to_series().diff().fillna(0) * HOP_S).cumsum()
        # Aproximación: tomamos los primeros K donde K*HOP_S <= first_n_s
        k = max(5, int(first_n_s / HOP_S))   # mínimo 5 ventanas
        df_base = df_spk.head(k)
        label = f"primeros {first_n_s:.0f}s"

    elif strategy == 'iqr':
        # Mediana ± 1.5*IQR para excluir outliers
        df_base = df_spk.copy()
        for col in FEATURE_COLS:
            q1, q3 = df_base[col].quantile([0.25, 0.75])
            iqr = q3 - q1
            df_base = df_base[df_base[col].between(q1 - 1.5*iqr, q3 + 1.5*iqr)]
        label = "IQR (toda la llamada)"

    baseline = {}
    for col in FEATURE_COLS:
        valid = df_base[col].dropna()
        if len(valid) < 3:
            print(f"  ADVERTENCIA: muy pocas muestras para '{col}' en {speaker}")
        baseline[col] = {
            'mean': valid.mean() if len(valid) > 0 else np.nan,
            'std':  max(valid.std(), 1e-6) if len(valid) > 1 else 1.0  # evitar div/0
        }

    print(f"{speaker} — base calculada con estrategia '{label}' ({len(df_base)} ventanas)")
    return baseline


baselines = {}
for spk in speakers:
    baselines[spk] = compute_baseline(df_wins, spk, BASELINE_STRATEGY, BASELINE_FIRST_N_S)

# Mostrar tabla de referencia
rows = []
for spk, base in baselines.items():
    for feat, stats in base.items():
        rows.append({'speaker': spk, 'feature': feat,
                     'base_mean': round(stats['mean'], 4),
                     'base_std':  round(stats['std'], 4)})
pd.DataFrame(rows).set_index(['speaker','feature'])
````

## 6. Score de desviación emocional

Para cada ventana calculamos:

$$z_i(t) = \frac{f_i(t) - \mu_i^{base}}{\sigma_i^{base}}$$

Donde $f_i(t)$ es la feature $i$ en el tiempo $t$, y $(\mu_i^{base}, \sigma_i^{base})$ son los parámetros de la base.

El **score compuesto** es una suma ponderada de z-scores absolutos:

$$S(t) = \sum_i w_i \cdot |z_i(t)|$$

Un $S(t)$ alto indica que varios features se alejaron simultáneamente de la base → probable activación emocional.

La **dirección emocional** (enojo vs. tristeza) se infiere del signo de los z-scores individuales.

````python
def compute_deviation_scores(df_wins: pd.DataFrame, baselines: dict,
                              weights: dict) -> pd.DataFrame:
    """
    Añade columnas z_<feature> y 'emotion_score' a df_wins.
    También añade 'direction': +1 = activación alta (enojo/excitación),
                               -1 = activación baja (tristeza/calma)
    """
    df = df_wins.copy()

    for spk in speakers:
        mask = df['speaker'] == spk
        base = baselines[spk]

        for feat, w in weights.items():
            if feat not in df.columns:
                continue
            mu  = base[feat]['mean']
            std = base[feat]['std']
            df.loc[mask, f'z_{feat}'] = (df.loc[mask, feat] - mu) / std

        # Score compuesto: suma ponderada de |z|
        z_cols = [f'z_{f}' for f in weights if f in df.columns]
        w_arr  = np.array([weights[f] for f in weights if f in df.columns])
        z_mat  = df.loc[mask, z_cols].values
        df.loc[mask, 'emotion_score'] = np.nansum(np.abs(z_mat) * w_arr, axis=1)

        # Dirección: pitch + energía determinan si es activación alta o baja
        dir_signal = df.loc[mask, 'z_pitch_mean'].fillna(0) + \
                     df.loc[mask, 'z_energy_db'].fillna(0)
        df.loc[mask, 'direction'] = np.sign(dir_signal)

    # Suavizado del score (ventana de 5 puntos) para reducir ruido
    for spk in speakers:
        mask = df['speaker'] == spk
        df.loc[mask, 'emotion_score_smooth'] = uniform_filter1d(
            df.loc[mask, 'emotion_score'].fillna(0), size=5
        )

    return df


df_wins = compute_deviation_scores(df_wins, baselines, FEATURE_WEIGHTS)

# Vista previa de scores
cols_preview = ['time_s', 'speaker', 'pitch_mean', 'energy_db',
                'z_pitch_mean', 'z_energy_db', 'emotion_score', 'direction']
df_wins[cols_preview].dropna().head(10)
````

## 7. Detección de momentos con activación emocional

Umbral adaptativo: un momento es "activo" si su score supera `ALERT_THRESHOLD` desviaciones estándar sobre la media del score de ese hablante.

Agrupamos ventanas consecutivas para obtener **eventos** (inicio, fin, duración, intensidad).

````python
def detect_emotional_events(df_wins: pd.DataFrame, threshold_sd: float) -> pd.DataFrame:
    """
    Marca ventanas con alta desviación y agrupa en eventos continuos.
    Devuelve DataFrame de eventos con [speaker, start_s, end_s, duration_s,
                                        peak_score, mean_direction].
    """
    df = df_wins.copy()
    events = []

    for spk in speakers:
        mask = df['speaker'] == spk
        df_spk = df[mask].copy().sort_values('time_s')

        # Umbral individual por hablante (media + N*std del score)
        s_mean = df_spk['emotion_score_smooth'].mean()
        s_std  = df_spk['emotion_score_smooth'].std()
        threshold = s_mean + threshold_sd * s_std

        df_spk['alert'] = df_spk['emotion_score_smooth'] > threshold
        df.loc[mask, 'alert']     = df_spk['alert'].values
        df.loc[mask, 'threshold'] = threshold

        print(f"{spk}: umbral = {threshold:.2f} "
              f"(media {s_mean:.2f} + {threshold_sd}*std {s_std:.2f})")

        # Agrupar ventanas consecutivas en eventos
        alert_rows = df_spk[df_spk['alert']].copy()
        if alert_rows.empty:
            print(f"  → Sin eventos detectados para {spk}")
            continue

        # Agrupar por continuidad temporal (gap < 2*HOP_S)
        alert_rows = alert_rows.sort_values('time_s')
        alert_rows['group'] = (
            (alert_rows['time_s'].diff() > 2 * HOP_S).cumsum()
        )

        for _, grp in alert_rows.groupby('group'):
            events.append({
                'speaker':        spk,
                'start_s':        grp['time_s'].min(),
                'end_s':          grp['time_s'].max() + WINDOW_S,
                'duration_s':     grp['time_s'].max() - grp['time_s'].min() + WINDOW_S,
                'peak_score':     grp['emotion_score_smooth'].max(),
                'mean_direction': grp['direction'].mean(),
            })

        print(f"  → {len(alert_rows.groupby('group'))} evento(s) detectado(s)")

    df_events = pd.DataFrame(events)
    if not df_events.empty:
        # Interpretar dirección
        df_events['tipo'] = df_events['mean_direction'].apply(
            lambda d: 'Alta activación (enojo/excitación)' if d > 0 else 'Baja activación (calma/tristeza)'
        )

    return df, df_events


df_wins, df_events = detect_emotional_events(df_wins, ALERT_THRESHOLD)

print("\n=== Eventos detectados ===")
if not df_events.empty:
    display(df_events.round(2))
else:
    print("No se detectaron eventos con el umbral actual. Prueba reducir ALERT_THRESHOLD.")
````

## 8. Visualización del arco emocional

El gráfico principal muestra, para cada hablante:
- La forma de onda coloreada por hablante
- El score emocional a lo largo del tiempo
- El umbral de alerta
- Las zonas destacadas en rojo donde se superó el umbral

````python
def plot_emotional_arc(y_full, sr, df_wins, df_events, df_segments):
    n_spk = len(speakers)
    fig = plt.figure(figsize=(18, 4 + 4 * n_spk))
    gs  = gridspec.GridSpec(1 + n_spk, 1, figure=fig, hspace=0.45)

    # ── Forma de onda global coloreada por hablante ────────────
    ax_wave = fig.add_subplot(gs[0])
    t_axis  = np.linspace(0, len(y_full) / sr, len(y_full))
    ax_wave.plot(t_axis, y_full, color='#CCCCCC', linewidth=0.3, zorder=1)

    for _, row in df_segments.iterrows():
        mask = (t_axis >= row['start']) & (t_axis <= row['end'])
        ax_wave.fill_between(t_axis, y_full, where=mask,
                             color=color_map[row['speaker']], alpha=0.4, zorder=2)

    # Marcar eventos como rectángulos rojos
    for _, ev in df_events.iterrows():
        ax_wave.axvspan(ev['start_s'], ev['end_s'],
                        color='red', alpha=0.15, zorder=3)

    patches = [mpatches.Patch(color=color_map[s], label=s) for s in speakers]
    patches.append(mpatches.Patch(color='red', alpha=0.3, label='Evento emocional'))
    ax_wave.legend(handles=patches, loc='upper right', fontsize=9)
    ax_wave.set_title('Forma de onda — Llamada completa', fontsize=12)
    ax_wave.set_xlabel('Tiempo (s)')
    ax_wave.set_ylabel('Amplitud')

    # ── Score emocional por hablante ───────────────────────────
    for i, spk in enumerate(speakers):
        ax = fig.add_subplot(gs[i + 1])
        df_spk = df_wins[df_wins['speaker'] == spk].sort_values('time_s')

        if df_spk.empty:
            ax.set_title(f'{spk} — sin datos')
            continue

        t   = df_spk['time_s'].values
        sc  = df_spk['emotion_score_smooth'].values
        thr = df_spk['threshold'].iloc[0] if 'threshold' in df_spk.columns else np.nan
        dir_vals = df_spk['direction'].values

        # Línea del score
        ax.plot(t, sc, color=color_map[spk], linewidth=1.8, label='Score emocional', zorder=3)

        # Colorear área según dirección
        ax.fill_between(t, sc, where=(dir_vals > 0) & (sc > 0),
                        color='#FF5722', alpha=0.2, label='Alta activación')
        ax.fill_between(t, sc, where=(dir_vals <= 0) & (sc > 0),
                        color='#607D8B', alpha=0.2, label='Baja activación')

        # Línea base y umbral
        ax.axhline(df_spk['emotion_score_smooth'].mean(),
                   color='green', linestyle='--', linewidth=1.2, alpha=0.7, label='Media base')
        if np.isfinite(thr):
            ax.axhline(thr, color='red', linestyle=':', linewidth=1.5,
                       alpha=0.9, label=f'Umbral ({ALERT_THRESHOLD}σ)')

        # Sombrear eventos
        for _, ev in df_events[df_events['speaker'] == spk].iterrows():
            ax.axvspan(ev['start_s'], ev['end_s'], color='red', alpha=0.12, zorder=1)
            ax.annotate('⚡', xy=((ev['start_s'] + ev['end_s']) / 2, thr),
                        fontsize=14, ha='center', va='bottom', color='red')

        ax.set_title(f'Arco emocional — {spk}', fontsize=12)
        ax.set_xlabel('Tiempo de llamada (s)')
        ax.set_ylabel('Score de desviación')
        ax.legend(loc='upper right', fontsize=8, ncol=2)

    plt.suptitle('Detección de Desviaciones Emocionales', fontsize=14, y=1.01, fontweight='bold')
    plt.savefig(OUTPUT_DIR / 'emotional_arc.png', dpi=150, bbox_inches='tight')
    plt.show()


plot_emotional_arc(y_full, sr, df_wins, df_events, df_segments)
````

## 9. Detalle de cada evento detectado

Para cada evento, visualizamos el pitch y la energía en ese fragmento concreto,
comparados con la base del hablante.

````python
def plot_event_detail(ev: pd.Series, y_full: np.ndarray, sr: int,
                      baselines: dict, margin_s: float = 3.0):
    """
    Muestra el detalle acústico de un evento emocional con contexto.
    """
    spk     = ev['speaker']
    t_start = max(0, ev['start_s'] - margin_s)
    t_end   = min(len(y_full) / sr, ev['end_s'] + margin_s)

    y_ev = y_full[int(t_start * sr): int(t_end * sr)]
    t_ax = np.linspace(t_start, t_end, len(y_ev))

    f0, voiced, _ = librosa.pyin(y_ev, fmin=75, fmax=500, sr=sr)
    t_f0 = np.linspace(t_start, t_end, len(f0))
    rms  = librosa.feature.rms(y=y_ev)[0]
    t_rms = np.linspace(t_start, t_end, len(rms))

    base_pitch  = baselines[spk]['pitch_mean']['mean']
    base_energy = baselines[spk]['energy_db']['mean']

    fig, axes = plt.subplots(3, 1, figsize=(14, 8), sharex=True)

    # Forma de onda
    axes[0].plot(t_ax, y_ev, color=color_map[spk], linewidth=0.5)
    axes[0].axvspan(ev['start_s'], ev['end_s'], color='red', alpha=0.15)
    axes[0].set_ylabel('Amplitud')
    axes[0].set_title(f'Evento emocional — {spk} '
                      f'[{ev["start_s"]:.1f}s – {ev["end_s"]:.1f}s]  '
                      f'Tipo: {ev["tipo"]}')

    # Pitch
    axes[1].plot(t_f0, f0, color='navy', linewidth=1.5, label='F0')
    axes[1].axhline(base_pitch, color='green', linestyle='--',
                    linewidth=1.5, label=f'Base: {base_pitch:.1f} Hz')
    axes[1].axvspan(ev['start_s'], ev['end_s'], color='red', alpha=0.1)
    axes[1].set_ylabel('Pitch (Hz)')
    axes[1].legend(fontsize=9)

    # Energía
    energy_db = librosa.amplitude_to_db(rms, ref=1.0)
    axes[2].fill_between(t_rms, energy_db, alpha=0.7,
                         color=color_map[spk], label='Energía RMS')
    axes[2].axhline(base_energy, color='green', linestyle='--',
                    linewidth=1.5, label=f'Base: {base_energy:.1f} dB')
    axes[2].axvspan(ev['start_s'], ev['end_s'], color='red', alpha=0.1)
    axes[2].set_ylabel('Energía (dB)')
    axes[2].set_xlabel('Tiempo (s)')
    axes[2].legend(fontsize=9)

    plt.tight_layout()
    fname = OUTPUT_DIR / f'evento_{spk}_{ev["start_s"]:.0f}s.png'
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()


if not df_events.empty:
    for _, ev in df_events.iterrows():
        plot_event_detail(ev, y_full, sr, baselines)
else:
    print("No hay eventos para detallar. Ajusta ALERT_THRESHOLD.")
````

## 10. Heatmap de z-scores — ¿qué features dispararon el evento?

````python
def plot_zscore_heatmap(df_wins: pd.DataFrame):
    """
    Mapa de calor: filas = tiempo, columnas = features, color = z-score.
    Permite ver qué característica causó cada desvío.
    """
    z_cols = [f'z_{f}' for f in FEATURE_WEIGHTS if f'z_{f}' in df_wins.columns]
    labels = list(FEATURE_WEIGHTS.keys())

    for spk in speakers:
        df_spk = df_wins[df_wins['speaker'] == spk].sort_values('time_s')
        if df_spk.empty:
            continue

        z_matrix = df_spk[z_cols].values.T   # (n_features, n_time)
        t_vals   = df_spk['time_s'].values

        # Recortar z-scores extremos para mejor visualización
        z_matrix = np.clip(z_matrix, -4, 4)

        fig, ax = plt.subplots(figsize=(16, 4))
        cmap = LinearSegmentedColormap.from_list(
            'emotion', ['#1565C0', '#FFFFFF', '#C62828']  # azul → blanco → rojo
        )

        im = ax.imshow(z_matrix, aspect='auto', cmap=cmap, vmin=-4, vmax=4,
                       extent=[t_vals[0], t_vals[-1], len(z_cols) - 0.5, -0.5])

        # Marcar eventos
        for _, ev in df_events[df_events['speaker'] == spk].iterrows():
            ax.axvline(ev['start_s'], color='black', linewidth=1.5, linestyle='--', alpha=0.7)
            ax.axvline(ev['end_s'],   color='black', linewidth=1.5, linestyle='--', alpha=0.7)

        ax.set_yticks(range(len(labels)))
        ax.set_yticklabels(labels)
        ax.set_xlabel('Tiempo (s)')
        ax.set_title(f'Z-scores por feature — {spk}  '
                     f'(rojo = alta activación, azul = baja)')
        plt.colorbar(im, ax=ax, label='z-score')
        plt.tight_layout()
        plt.savefig(OUTPUT_DIR / f'zscore_heatmap_{spk}.png', dpi=150, bbox_inches='tight')
        plt.show()


plot_zscore_heatmap(df_wins)
````

## 11. Resumen ejecutivo por hablante

````python
def print_executive_summary(df_events, df_segments, baselines):
    total_duration = len(y_full) / sr
    print("=" * 60)
    print(f"RESUMEN EMOCIONAL DE LA LLAMADA")
    print(f"Duración total: {total_duration:.1f} s ({total_duration/60:.1f} min)")
    print("=" * 60)

    for spk in speakers:
        total_speech = df_segments[df_segments['speaker'] == spk]['duration'].sum()
        n_events = len(df_events[df_events['speaker'] == spk]) if not df_events.empty else 0

        print(f"\n{'─'*40}")
        print(f"Hablante: {spk}")
        print(f"  Tiempo total de habla : {total_speech:.1f} s ({total_speech/60:.1f} min)")
        print(f"  Pitch base            : {baselines[spk]['pitch_mean']['mean']:.1f} Hz")
        print(f"  Energía base          : {baselines[spk]['energy_db']['mean']:.1f} dB")
        print(f"  Eventos detectados    : {n_events}")

        if n_events > 0 and not df_events.empty:
            spk_events = df_events[df_events['speaker'] == spk]
            for j, (_, ev) in enumerate(spk_events.iterrows(), 1):
                print(f"\n  Evento #{j}:")
                print(f"    Tiempo     : {ev['start_s']:.1f}s – {ev['end_s']:.1f}s "
                      f"(duración: {ev['duration_s']:.1f}s)")
                print(f"    Intensidad : {ev['peak_score']:.2f} (score pico)")
                print(f"    Tipo       : {ev['tipo']}")

    print(f"\n{'='*60}")
    print(f"Archivos generados en: {OUTPUT_DIR}/")


print_executive_summary(df_events, df_segments, baselines)

# Exportar
df_wins.to_csv(OUTPUT_DIR / 'windows_with_scores.csv', index=False)
if not df_events.empty:
    df_events.to_csv(OUTPUT_DIR / 'emotional_events.csv', index=False)
print("\nCSVs exportados.")
````

## 12. Guía de ajuste y próximos pasos

### Parámetros clave a ajustar

| Parámetro | Efecto | Cuándo cambiar |
|---|---|---|
| `WINDOW_S` | Ventana más grande → menos ruido, menos resolución temporal | Si los eventos son muy cortos, reducir a 2-3s |
| `BASELINE_STRATEGY = 'iqr'` | Base más robusta si hay emoción al inicio | Si la persona llega ya enojada al llamar |
| `BASELINE_FIRST_N_S` | Más segundos → base más estable | Ampliar si el inicio también tiene emoción |
| `ALERT_THRESHOLD` | Más alto → menos alertas (solo picos claros) | Reducir a 1.5 si te pierdes eventos; subir a 2.5 para reducir falsos positivos |
| `FEATURE_WEIGHTS` | Cambiar qué features tienen más peso | Priorizar pitch para enojo; energía para agresividad |

### Limitaciones actuales
- La base calculada en la misma llamada asume que la mayor parte es neutral.
- Jitter/Shimmer son sensibles al ruido de canal telefónico — interpretar con cautela.
- `speech_rate` via onset detection es una aproximación; para mayor precisión usar ASR.

### Próximos pasos para producción
1. **Base de referencia externa**: grabar a cada agente en una sesión neutral y usar esa base.
2. **Clasificador supervisado**: etiquetar algunos momentos y entrenar un modelo (SVM, random forest) sobre los z-scores.
3. **Integrar ASR**: correlacionar los eventos acústicos con las palabras pronunciadas.
4. **Métricas de negocio**: cruzar eventos con resultado de la llamada (conversión, satisfacción).
